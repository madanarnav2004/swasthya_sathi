```python
from transformers import AutoTokenizer, AutoModel
```

```python
tokenizer = AutoTokenizer.from_pretrained("THUDM/chatglm2-6b-int4", trust_remote_code=True)
```

```python
model = AutoModel.from_pretrained("THUDM/chatglm2-6b-int4", trust_remote_code=True).half().cuda()
```

```python
from transformers import MarianMTModel, MarianTokenizer
```

```python
def translate_to_hindi(text):
  model_name = "Helsinki-NLP/opus-mt-en-hi"
  tokenizer = MarianTokenizer.from_pretrained(model_name)
  model = MarianMTModel.from_pretrained(model_name)


  max_length = 512
  chunks = [text[i:i+max_length] for i in range(0, len(text), max_length)]

  translated_chunks = []
  for chunk in chunks:
      inputs = tokenizer(chunk, return_tensors="pt")
      outputs = model.generate(**inputs)
      decoded_text = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]
      translated_chunks.append(decoded_text)

  hindi_text = " ".join(translated_chunks).strip()
  return hindi_text

english_text = response
hindi_text = translate_to_hindi(english_text)
print(f"English: {english_text}")
print(f"Hindi: {hindi_text}")

```
